```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('england-premier-league-2007-to-2008.csv')
df1 = pd.read_csv('england-premier-league-2008-to-2009.csv')
df2 = pd.read_csv('england-premier-league-2009-to-2010.csv')
df3 = pd.read_csv('england-premier-league-2011-to-2012.csv')
df4 = pd.read_csv('england-premier-league-2012-to-2013.csv')
df5 = pd.read_csv('england-premier-league-2013-to-2014.csv')
df6 = pd.read_csv('england-premier-league-2014-to-2015.csv')
df7 = pd.read_csv('england-premier-league-2015-to-2016.csv')
df8 = pd.read_csv('england-premier-league-2016-to-2017.csv')
df9 = pd.read_csv('england-premier-league-2017-to-2018.csv')
df10 = pd.read_csv('england-premier-league-2018-to-2019.csv')
df11 = pd.read_csv('england-premier-league-2019-to-2020.csv')
df
```

```python
soccer = pd.concat([df, df1, df2,df3,df4,df5,df6,df7,df8,df9,df10,df11], axis=0)
soccer = soccer.loc[:,'Div':'AR']
print(soccer.isna().sum())
soccer
```
